```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


call= {10700:266,10800:189,10900:120,11000:67,11100:32,11200:13,11300:6.1}
put= {10700:43,10800:63,10900:93,11000:139,11100:207,11200:290,11300:360}
x = np.arange(10500,11501)

def callr(K):
    global x
    global call
    return np.maximum(x-K,0)-call[K]
def putr(K):   
    global x
    global call
    return np.maximum(K-x,0)-put[K]

y1= callr(10700)
y2= -callr(11100)
y3 = y1+y2
plt.plot(x,y1,'r',x,y2,'g',x,y3,'b',[x[0],x[-1]],[0,0],'--k')
plt.show()
```

```python
y7= putr(10700)
y8= -putr(11100)
y9 = y7+y8
plt.plot(x,y7,'r',x,y8,'g',x,y9,'b',[x[0],x[-1]],[0,0],'--k')
plt.show()
```
